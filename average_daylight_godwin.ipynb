```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
```

```python
daylight_data_df = pd.read_csv("Happiness_Cities_Weather_2018_clean.csv")

daylight_data_df.head(366)
```

```python
#copy data frame to prevent unwanted changes to clean data
daylight_null_drop = daylight_data_df.copy()
```

```python
sunrise_unix = []
sunset_unix = []
sunrise_converted = []
sunset_converted = []
total_daylight_hours = []

#append unix code to sunrise and sunset lists
for index, row in daylight_null_drop.iterrows():
    
    sunrise_unix.append(daylight_null_drop["Sunrise Time"][index])
    sunset_unix.append(daylight_null_drop["Sunset Time"][index])
    
#convert unix time, and append datetime datatype to sunrise list
for unix_time in sunrise_unix: 
    
    converted_time = timestamp = datetime.datetime.fromtimestamp(unix_time)
    sunrise_converted.append(converted_time)

#convert unix time, and append datetime datatype to sunset list
for unix_time in sunset_unix:
    
    converted_time = timestamp = datetime.datetime.fromtimestamp(unix_time)
    sunset_converted.append(converted_time)

#iterate through index's, and subtact sunset from sunrise to find total daylight hrs for day (datetype datatype)
for index in range(len(sunset_converted)): 
    
     total_daylight_hours.append(sunset_converted[index] - sunrise_converted[index]) 

#create new daylight hours column and append to dataframe
daylight_null_drop["Total Daylight Hours"] = total_daylight_hours

daylight_null_drop.head(365) 
```

```python
#compute total seconds per day and create new column on DataFrame containing total seconds
daylight_null_drop["Daylight_Seconds"] = [x.seconds for x in daylight_null_drop["Total Daylight Hours"]]
```

```python
#groupby total score and daylight seconds; find average seconds; reset index; order and title columns, and save df to variable
daylight_happiness= pd.DataFrame(daylight_null_drop.groupby("Total Score")["Daylight_Seconds"].mean()).reset_index()
daylight_happiness.columns = ["Total Score", "Avg_Secs"]
daylight_happiness.head()
```

```python
#find formatted average time per city and create new column and store to new column "Avg_Time"
daylight_happiness["Avg_Time"] = [datetime.timedelta(seconds=x) for x in daylight_happiness.Avg_Secs]
```

```python
#display first 20 rows
daylight_happiness.head(20)
```

```python
plt.figure(figsize=(20,10))

plt.title("Happiness Rating vs. Average Daylight (2018)", weight="bold", fontsize=20)
plt.xlabel("Happiness Rating", fontsize=16)
plt.ylabel("Average Daylight (hh:mm:ss)", fontsize=16)
plt.grid(b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.xlim(25, 75)

plt.scatter(daylight_happiness["Total Score"], daylight_happiness.Avg_Secs)
plt.yticks(plt.yticks()[0], [datetime.timedelta(seconds=x) for x in plt.yticks()[0]])

#save figure as .png
plt.savefig("avg_daylight_scatterPlot.png")

plt.show()
```
