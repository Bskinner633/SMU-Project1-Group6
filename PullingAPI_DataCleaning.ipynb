```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
```

```python
happiness_cities = pd.read_csv("Happiest_Cities_consolidated.csv")
happiness_cities.head()
```

```python
happiness_cities["Lat"] = ""
happiness_cities["Lng"] = ""
happiness_cities.head()
```

```python
# Google API Key
from config import gkey
```

```python
params = {"key": gkey}

for index, row in happiness_cities.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    city = row['City']
    state = row['State']

    # update address key value
    params['address'] = f"{city},{state}"

    # make request
    cities_lat_lng = requests.get(base_url, params=params)
    
    # print the cities_lat_lng url, avoid doing for public github repos in order to avoid exposing key
    # print(cities_lat_lng.url)
    
    # convert to json
    cities_lat_lng = cities_lat_lng.json()

    happiness_cities.loc[index, "Lat"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]
    happiness_cities.loc[index, "Lng"] = cities_lat_lng["results"][0]["geometry"]["location"]["lng"]

# Visualize to confirm lat lng appear
happiness_cities
```

```python
happiness_cities.to_csv("Happiness_Cities_withLL.csv", index=False)
```

```python
from datetime import datetime
import time
```

```python
datelist = pd.date_range(datetime(2018, 11, 2), datetime(2018, 12, 31)).tolist()

dateUTC = [int(time.mktime(t.timetuple())) for t in datelist]

dateUTC
```

```python
happiness_cities_full = happiness_cities.copy()

for x in range(len(dateUTC) -1):
    happiness_cities_full = happiness_cities_full.append(happiness_cities.copy())
    
    
```

```python
happiness_cities_full = happiness_cities_full.sort_values("City").reset_index(drop=True)

happiness_cities_full.head(100)
```

```python
happiness_cities_full["DateUTC"] = dateUTC * 182
```

```python
happiness_cities_full.head(100)
```

```python
happiness_cities_full["Summary"] = ""
happiness_cities_full["Icon"] = ""
happiness_cities_full["Sunrise Time"]= ""
happiness_cities_full["Sunset Time"] = ""
happiness_cities_full["Cloud Cover"] = ""
happiness_cities_full["UV Index"] = ""
happiness_cities_full.head()
```

```python
from config import darkskykey
```

```python
url = "https://api.darksky.net/forecast"



for index, row in happiness_cities_full.iterrows():
    #     get lat, lng from df
    lat = row["Lat"]
    lng = row["Lng"]
    time = row["DateUTC"]
    query_url = f"{url}/{darkskykey}/{lat},{lng},{time}?"
    data = requests.get(query_url).json()
 
    try:  
        happiness_cities_full.loc[index, "Summary"] = (data["daily"]["data"][0]["summary"])
        happiness_cities_full.loc[index, "Icon"] = (data["daily"]["data"][0]["icon"])
        happiness_cities_full.loc[index, "Sunrise Time"] = (data["daily"]["data"][0]["sunriseTime"])
        happiness_cities_full.loc[index, "Sunset Time"] = (data["daily"]["data"][0]["sunsetTime"])
        happiness_cities_full.loc[index, "Cloud Cover"] = (data["daily"]["data"][0]["cloudCover"])
        happiness_cities_full.loc[index, "UV Index"] = (data["daily"]["data"][0]["uvIndex"])
    except KeyError: 
        print("Not Available")

    
```

```python
happiness_cities_full.to_csv("Happiness_Cities_Weather_Nov-Dec.csv", index=False)
```
